```python
from transformers import pipeline


```

```python
sentiment_analyzer = pipeline("sentiment-analysis")


```

```python
text = "The product quality is excellent and I am very happy with the service."

```

```python
result = sentiment_analyzer(text)
print(result)
```

```python
def analyze_sentiment(text):
    return sentiment_analyzer(text)
```

```python
analyze_sentiment("The delivery was late and the product is damaged.")
```

## Observations

- The sentiment analysis model successfully classified customer reviews into **Positive** and **Negative** categories.
- Reviews with clear emotional expressions produced **high confidence scores**, indicating strong prediction reliability.
- Mixed or neutral feedback resulted in **lower confidence scores**, reflecting realistic ambiguity in language.
- The pretrained **DistilBERT** model performed well without requiring any additional training or fine-tuning.
- The Hugging Face **sentiment-analysis pipeline** simplified model loading and inference.
- The system can process **multiple customer reviews**, making it suitable for analyzing large datasets.
- Overall, the project demonstrates an effective application of Transformer-based models for customer feedback analysis.
